```python
import pandas as pd
import requests
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the sire to Scrape 
url = 'https://www.winemag.com/region/oregon/'
browser.visit(url)
page = requests.get(url)
```

```python
html = browser.html
soup = soup(html, 'html.parser')
```

```python
lists = soup.find_all('section', class_="wine-reviews")

for list in lists:
    title = list.find_all('h3', class_= "title")
    location = list.find_all('span', class_= "appellation")
    review = list.find_all('div', class_= "excerpt")
    points = list.find_all('span', class_= "rating")
    prices = list.find_all('span', class_= "price")
```

```python
table = [title, location, review, points, prices ]
print(table)
```

```python
table_df = pd.DataFrame(table)

table_df
```

```python
new_table_df = table_df.transpose()
new_table_df
```
